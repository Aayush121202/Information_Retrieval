```python
import pyterrier as pt
```

```python
pt.init()
```

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_json('/content/lab1-data.json')
```

```python
df.head()
```

```python
df.rename(columns={"id":"docno"}, inplace=True)
```

```python
df["docno"] = df["docno"].astype(str)
```

```python
df.shape
```

```python
df= df.dropna()
```

```python
df.shape
```

```python
df_subset = df[["text", "docno"]]
dictionary = df_subset.to_dict(orient="records")
```

```python
indexer = pt.IterDictIndexer("./index", overwrite=True)
indexref = indexer.index(dictionary)
```

```python
ind= pt.IndexFactory.of(indexref)
```

```python
print(ind.getCollectionStatistics().toString())
```

```python
tf_idf = pt.terrier.Retriever(ind, wmodel="TF_IDF")
```

```python
tf_idf
```

```python
results = tf_idf.search("barbie")
```
